### pca

```python
pcadata = preprocessing.scale(result)
pca = PCA(n_components=46)

principalComp = pca.fit_transform(pcadata)

pDF = pd.DataFrame(data = principalComp, columns =result.columns)
pDF.head()
```

```python
load = pca.components_
num = pca.n_features_
pc = ["PC"+str(i) for i in list(range(1, num+1))]
load_df = pd.DataFrame.from_dict(dict(zip(pc, load)))
load_df['feature'] = result.columns.values
load_df = load_df.set_index('feature')
load_df
```

```python
pca.explained_variance_ratio_
```

```python
plt.scatter(principalComp[:, 0], principalComp[:, 1], c='grey')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Principal Component Scores')
plt.show()
```

```python
loadings = pca.components_
cluster.biplot(cscore=principalComp, loadings=load, labels=load_df.columns.values, 
               var1=round(pca.explained_variance_ratio_[0]*100, 2),
    var2=round(pca.explained_variance_ratio_[1]*100, 2),show=True,dim=(15,10))
```

### KMEANS

```python
K = range(2,12)
wcss=[]
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(result)
    wcss.append(kmeanModel.inertia_)
plt.figure(figsize=(7,5))
plt.plot(range(2,12),wcss)
plt.xlabel("number of k (cluster) value")
plt.ylabel("wcss")
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=0)
kmeans.fit(result)
print('No. of iterations: ', kmeans.n_iter_)
```

### DBSCAN

```python
re = result.copy()
for i in range(1,12):
    for j in range(1,12):
        dbscan_cluster = DBSCAN(eps=i*0.5, min_samples=j)
        clusters=dbscan_cluster.fit_predict(result)
        if len(np.unique(clusters))>=2:
            dbresult =re.append({'Eps':i*0.5,'Min_Samples':j,'Numberof  Cluster':len(np.unique(clusters)),'SilhouetteScore':silhouette_score(re,clusters),'Davies Bouldin Score':davies_bouldin_score(re,clusters)},ignore_index=True)
```
